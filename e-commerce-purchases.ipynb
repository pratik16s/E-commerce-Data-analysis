# What are the top 5 most popular email providers

```python
# create new list
# split only '' provider name '' after @ (index 1)
# append on list
# create new column with this new values
# then, value_counts() : count by provider
```

```python
list = []
for email in data['Email']:
   
    list.append(email.split('@')[1])
```

```python
data['Email Provider'] = list
```

```python
data
```

```python
data['Email Provider'].value_counts()
```

```python
data['Email Provider'].value_counts().head(5)
```

#### Lambda method

```python
data['Email'].apply(lambda x:x.split('@')[1]).value_counts().head()
```
